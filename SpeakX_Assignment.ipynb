```python
#Importing required libraries
import numpy as np
import pandas as pd
```

**DATA COLLECTION & PREPROCESSING**

```python
#Loading the Data
df=pd.read_csv("/content/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
#Checking Information of the Dataset
df.info()
```

```python
#Displaying First Five Rows of Dataset
df.head()
```

```python
#Checking Shape of Dataset
df.shape
```

```python
#Checking For Null Values in each Column
df.isnull().sum()
```

```python
#Dropping CustomerId Column, as it plays no role
df=df.drop(columns=["customerID"])
```

```python
#Displaying First Five Rows of Dataset
df.head()
```

```python
#Checking For Duplicates
df.duplicated().sum()
```

```python
#Dropping Duplicates
df=df.drop_duplicates()
```

```python
#Displaying Information of the Dataset
df.info()
```

**FEATURE ENGINEERING & FEATURE SCALING**

```python
#Seperating the Churn Column & Checking the Shape
target=df1["Churn"]
target.shape
```

```python
#Dropping Churn Column from the Dataset
df1=df1.drop(columns=["Churn"])
df1.shape
```

```python
#Partitioning the Dataset into Training & Testing Data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(df1, target, test_size=0.3)
```

```python
#Checking the shapes after partition
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Describing the Dataset to check for Scaling
x_train.describe()
```

```python
#Scaling the Dataset
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.fit_transform(x_test)
x_train_sc=pd.DataFrame(x_train_sc, columns=x_train.columns)
x_test_sc=pd.DataFrame(x_test_sc, columns=x_test.columns)
```

```python
x_train_sc.describe()
```

**MODEL BUILDING & EVALUATION**

```python
#Creating Lists to store the Accuracies of each Algorithm to evaluate them later
algo=["LOGISTIC", "SVM", "DECISION TREE", "KNN", "RANDOM FOREST", "ADABOOST", "MLP"]
sc_train_acc=[]
sc_test_acc=[]
```

```python
#Implementing Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train_sc, y_train)
```

```python
#Predicting the Results
x_train_pred=lr.predict(x_train_sc)
x_test_pred=lr.predict(x_test_sc)
```

```python
#Evaluating the Model
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
print("Training Accuracy:", accuracy_score(x_train_pred, y_train))
print("Testing Accuracy:", accuracy_score(x_test_pred, y_test))
print("Confusion Matrix:")
print(confusion_matrix(x_train_pred, y_train))
print("Precision Score:", precision_score(x_train_pred, y_train))
print("Recall Score:", recall_score(x_train_pred, y_train))
print("F1 Score:", f1_score(x_train_pred, y_train))

sc_train_acc.append(accuracy_score(x_train_pred, y_train))
sc_test_acc.append(accuracy_score(x_test_pred, y_test))
```
